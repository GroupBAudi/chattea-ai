````python
import pandas as pd

# Load dataset
df = pd.read_csv("chattea.csv")

# Count samples per intent
intent_counts = df['intent'].value_counts()

# Show intents with only 1 sample
rare_intents = intent_counts[intent_counts == 1]
print("Intents with only 1 sample:")
print(rare_intents)
````

````python
import pandas as pd

df = pd.read_csv("chattea.csv")

# Count samples per intent
intent_counts = df['intent'].value_counts().reset_index()
intent_counts.columns = ['intent', 'count']

# Show all rows without truncation
pd.set_option('display.max_rows', None)

print(intent_counts)
````
